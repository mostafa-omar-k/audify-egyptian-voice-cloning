```python
import os
import torch
import torchaudio
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts
```

```python
from TTS.api import TTS
```

```python
%cd /kaggle/working
```

```python
SPEAKER_REFERENCE = []
for i in range(1,240):
    if i in [66,64,95]:
        continue
    else:
        wav = f"/kaggle/input/dataset/{i}.wav"
        SPEAKER_REFERENCE.append(wav)
```

```python
print(SPEAKER_REFERENCE)
```

```python

# Add here the xtts_config path
CONFIG_PATH = "/kaggle/working/config.json"
# Add here the vocab file that you have used to train the model
TOKENIZER_PATH = "/kaggle/working/vocab.json"
# Add here the checkpoint that you want to do inference with
XTTS_CHECKPOINT = "/kaggle/working/best_model.pth"
# Add here the speaker reference
# SPEAKER_REFERENCE = ["/kaggle/input/test-data/MN.wav"]

# output wav path
OUTPUT_WAV_PATH = "/kaggle/working/output.wav"

print("Loading model...")
config = XttsConfig()
config.load_json(CONFIG_PATH)
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_path=XTTS_CHECKPOINT, vocab_path=TOKENIZER_PATH, use_deepspeed=False)
# model.cuda()
print("Model loaded...")
```

```python
print("Computing speaker latents...")
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=SPEAKER_REFERENCE)
```

```python
##مش عايزين في الهيصة دي نتسحل وننسى ان في مشروع لسه متعملش فيه حاgة
import time
print("Inference...")
t0 = time.time()
chunks = model.inference_stream(
    "المهم عشان مطولش عليكم العودة إن شاء الله هتبءى النهارده الساعة سبعة",
    "ar",
    gpt_cond_latent,
    speaker_embedding,
    temperature=0.7,
    speed = 1.1
)

wav_chuncks = []
for i, chunk in enumerate(chunks):
    if i == 0:
        print(f"Time to first chunck: {time.time() - t0}")
    print(f"Received chunk {i} of audio length {chunk.shape[-1]}")
    wav_chuncks.append(chunk)
wav = torch.cat(wav_chuncks, dim=0)
torchaudio.save("xtts_streaming.wav", wav.squeeze().unsqueeze(0).cpu(), 24000)
print("Downloaded")
```

```python
print("Inference...")
out = model.inference(
    "ولكن الظاهر ان مفيش حاgة نئدر نعملها",
    "ar",
    gpt_cond_latent,
    speaker_embedding,
    temperature=0.7,
    # Add custom parameters here
)
torchaudio.save(OUTPUT_WAV_PATH, torch.tensor(out["wav"]).unsqueeze(0), 24000)
print("Downloaded")
```

```python
from pedalboard import *
from pedalboard.io import AudioFile
```

```python
with AudioFile('/kaggle/working/output.wav').resampled_to(24000) as f:
    audio = f.read(f.frames)
```

```python
board = Pedalboard([
        NoiseGate(threshold_db=-30, ratio=1.5, release_ms=250),
        Compressor(threshold_db=12, ratio=2.5),
        LowShelfFilter(cutoff_frequency_hz=400, gain_db=5, q=1),
        Gain(gain_db=0)
    ])
    
result = board(audio, 24000)
```

```python
with AudioFile('processed-output.wav', 'w', 24000, result.shape[0]) as f:
      f.write(result)
```
